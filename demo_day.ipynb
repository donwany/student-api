```python
from pymongo import MongoClient
```

```python
# client = MongoClient("mongodb://127.0.0.1:27017")

client = MongoClient("mongodb://localhost:27017")
```

```python
# create a dababase

db = client["school"]

# create a collection
collection = db["students"]
```

## Insert data

```python
student = {
    "name": "John Doe",
    "age": 20,
    "grade": "A",
}

collection.insert_one(student)
```

## insert many documents

```python
# insert multiple students
students = [
    {"name": "Jane Doe", "age": 22, "grade": "B"},
    {"name": "Jim Doe", "age": 21, "grade": "C"},
    {"name": "Jack Doe", "age": 23, "grade": "A"},
]
collection.insert_many(students)
```

## Read from mongodb

```python
from pprint import pprint

# fetch all students
for student in collection.find():
    pprint(student)
```

```python
student = collection.find_one({"name": "Jim Doe"})
print(student)
```

```python
for student in collection.find({"grade": "C"}):
    pprint(student)
```

```python
collection.update_one({"name": "Jim Doe"},{"$set": {"grade": 'A+'}})
```

```python
collection.update_one({"name": "Jim Doe"},{"$push": {"courses": "AI"}})
```

```python
collection.delete_one({"name": "Jack Doe"})
```

## delete many

```python
collection.delete_many({"age": 22})
```
