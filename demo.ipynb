```python
import requests

class StudentAPIClient:
    def __init__(self, base_url: str):
        self.base_url = base_url

    # -------------------------
    # GET ALL STUDENTS
    # -------------------------
    def get_all_student(self):
        url = f"{self.base_url}/v1/students"

        response = requests.get(url)
        return response.json()

    # -------------------------
    # CREATE STUDENT
    # -------------------------
    def create_student(self, name: str, age: int, course: str):
        url = f"{self.base_url}/v1/students"
        payload = {
            "name": name,
            "age": age,
            "course": course
        }

        response = requests.post(url, json=payload)
        return response.json()

    # -------------------------
    # GET STUDENT BY ID
    # -------------------------
    def get_student(self, student_id: str):
        url = f"{self.base_url}/v1/students/{student_id}"

        response = requests.get(url)
        return response.json()

    # -------------------------
    # UPDATE STUDENT
    # -------------------------
    def update_student(self, student_id: str, name: str, age: int, course: str):
        url = f"{self.base_url}/v1/students/{student_id}"
        payload = {
            "name": name,
            "age": age,
            "course": course
        }

        response = requests.put(url, json=payload)
        return response.json()

    # -------------------------
    # DELETE STUDENT
    # -------------------------
    def delete_student(self, student_id: str):
        url = f"{self.base_url}/v1/students/{student_id}"

        response = requests.delete(url)
        return response.json()
```

### Usage

```python
BASE_URL = "https://student-api-jie2.onrender.com"

client = StudentAPIClient(base_url=BASE_URL)
```

```python
# CREATE
res = client.create_student(name="David", age=76, course="Agentic AI")
print(res)
```

```python
# GET
student_id = "69f159985d2f8ca983d80f8c"
print(client.get_student(student_id))
```

```python
# UPDATE
print(client.update_student(student_id, "Theo", 99, "Python"))
```

```python
# DELETE
student_id = "69f1408db5834ebcad854739"
print(client.delete_student(student_id))
```
