```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["customer_db"]       # Database
collection = db["customer"]    # Collection
```

## Insert document

```python
student = {
    "name": "Theo",
    "age": 26,
    "courses": ["AI", "Data Science"],
    "graduated": False
}

result = collection.insert_one(student)
print("Inserted ID:", result.inserted_id)
```

## Insert many

```python
customers = [
    {"name": "Alice", "age": 22, "courses": ["Math"], "graduated": False},
    {"name": "Bob", "age": 24, "courses": ["Physics"], "graduated": True}
]

collection.insert_many(customers)
```

## Read

```python
for customer in collection.find():
    print(customer)
```

## Find one Student

```python
customer = collection.find_one({"name": "Theo"})
print(customer)
```

## Query with one condition

```python
for customer in collection.find({"graduated": False}):
    print(customer)
```

## Update

```python
collection.update_one(
    {"name": "Theo"},
    {"$set": {"age": 27}}
)
```

## Add a course

```python
collection.update_one(
    {"name": "Theo"},
    {"$push": {"courses": "Machine Learning"}}
)
```

## Update multiple customers

```python
collection.update_many(
    {"graduated": False},
    {"$set": {"status": "active"}}
)
```

## Delete one student

```python
collection.delete_one({"name": "Bob"})
```

## Delete multiple students

```python
collection.delete_many({"graduated": True})
```
